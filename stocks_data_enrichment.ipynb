```python
import pandas as pd
import talib
```

```python
stock=pd.read_csv("Stocks.csv")
```

```python
stock
```

```python
stock['SMA_AAPL'] = talib.SMA(stock['AAPL'], timeperiod=14)
stock['SMA_MSI'] = talib.SMA(stock['MSI'], timeperiod=14)
stock['SMA_SBUX'] = talib.SMA(stock['SBUX'], timeperiod=14)

stock['EMA_AAPL'] = talib.EMA(stock['AAPL'], timeperiod=14)
stock['EMA_MSI'] = talib.EMA(stock['MSI'], timeperiod=14)
stock['EMA_SBUX'] = talib.EMA(stock['SBUX'], timeperiod=14)


stock['RSI_AAPL'] = talib.RSI(stock['AAPL'], timeperiod=14)
stock['RSI_MSI'] = talib.RSI(stock['MSI'], timeperiod=14)
stock['RSI_SBUX'] = talib.RSI(stock['SBUX'], timeperiod=14)
```

```python
stock
```

```python
stock.to_csv('Stocks2.csv', index=False)  
```
